# Time Series Data Application

## Importing packages

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

## Loading Dataset

```python
df = pd.read_csv('D15data1.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
type(df['Temp'])
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
plt.plot(df['Date'],df['Temp'])
```

```python
df['M12'] = df['Temp'].rolling(12).mean()
df
```

```python
df['Month'] = [i.month for i in df['Date']]
df
```

```python
df['Year'] = [i.year for i in df['Date']]
```

```python
df['Series'] = np.arange(1,len(df)+1)
df
```

```python
df.drop(df[['Date','M12']], axis=1,inplace=True)
df
```

```python
df = df[['Series','Year','Month','Temp']]
```
