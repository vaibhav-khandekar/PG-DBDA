## Imporitng packages

```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

## Loading dataset

```python
df = pd.read_csv('D14data3.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
transactions = []
for i in range(0,75000):
  transactions.append(str(df.values[i,j]for j in range(0,20)))
```

## Training model for Apriori

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3)
```

```python
res = list(rules)
```
